```python
def read_file(filename):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            return file.read()
    except UnicodeDecodeError:
        # Se a leitura com utf-8 falhar, tente com latin-1
        with open(filename, 'r', encoding='latin-1') as file:
            return file.read()
    except FileNotFoundError:
        print(f"Arquivo '{filename}' não encontrado.")
        return None
    except Exception as e:
        print(f"Erro ao ler o arquivo '{filename}': {str(e)}")
        return None

```

```python
def gera_amostra(lista, qtd_valores):
    # Se a quantidade de valores solicitada for maior do que o tamanho da lista - 1 (desconsiderando o cabeçalho), retorna a lista completa sem o cabeçalho
    if qtd_valores >= len(lista) - 1:
        return lista[1:], lista[0]

    # Remove o cabeçalho da lista para evitar duplicatas
    lista_sem_cabecalho = lista[1:]
    

    # Gera os índices aleatórios da amostra, excluindo o cabeçalho
    indices_amostra = random.sample(range(lista_sem_cabecalho), qtd_valores)
    print(indices_amostra)
    # Seleciona as linhas correspondentes aos índices escolhidos
    amostra = [lista_sem_cabecalho[i] for i in indices_amostra]

    return amostra, lista[0]
```

```python
def salva_csv (caminho, lista, cabecalho):

    # Salva a amostra em um novo arquivo CSV
    try:
        with open(caminho, 'w', newline='', encoding='utf-8') as arquivo:
            escritor_csv = csv.writer(arquivo)
            escritor_csv.writerow(cabecalho) 
            for linha in lista:
                escritor_csv.writerow(linha)

        print(f"Amostra de {len(lista)} valores salva com sucesso ")
    except Exception as erro:
        print(f'Ocorreu o seguinte erro na execução da função: {erro}')
```

```python
import csv
import random 
from datetime import datetime
from collections import Counter

#Camminho do arquivo de amostra que foi salvo recentemente 
caminho_arquivo_amostra = "amostra.csv"

#Variaveis para que os dados sejam adicionados conforme forem localizados
jogos_gratuitos = 0
jogos_pagos = 0

anos_ds_lancamentos = []

contagem_genero = {}
#Leitura do arquivo csv
with open(caminho_arquivo_amostra, 'r', encoding='utf-8') as arquivo_csv:
    leitor_csv = csv.reader(arquivo_csv)
    for linha in leitor_csv:
        # Pergunta 1: Verifique se o jogo é gratuito ou pago
        preco = linha[6]
        if preco == '0.0':
            jogos_gratuitos += 1
        if preco == 0.0:
            jogos_gratuitos += 1
        else:
            jogos_pagos += 1

        #Pergunta 2: Verificar o ano com mais jogos lançados
        dt_lancamento = linha[2]

        
        if len(dt_lancamento.split(", ")) == 2:
            ano_lancamento = dt_lancamento.split(", ")[1] 
            anos_ds_lancamentos.append(ano_lancamento)

        #Pergunta 3: Qual é a qquantidade de jogos para cada gêneros 
        genres = linha[35].split(",") 

        # Iterar sobre os gêneros e incrementar o contador para cada um
        for genre in genres:
            genre = genre.strip()  
            contagem_genero[genre] = contagem_genero.get(genre, 0) + 1

total_jogos = jogos_gratuitos + jogos_pagos

# Calcule o percentual de jogos gratuitos e pagos
percentual_gratuitos = (jogos_gratuitos / total_jogos) * 100
percentual_pagos = (jogos_pagos / total_jogos) * 100

print('PERGUNTA 1')
print(f'Percentual de jogos gratuitos: {percentual_gratuitos:.2f}%')
print(f'Percentual de jogos pagos: {percentual_pagos:.2f}%')


# Contar o número de lançamentos por ano
cont_ano = Counter(anos_ds_lancamentos)

# Encontrar o ano (ou anos em caso de empate) com o maior número de novos jogos
max_games = max(cont_ano.values())
ano_mais_jogos = [year for year, count in cont_ano.items() if count == max_games]

print('PERGUNTA 2')
if len(ano_mais_jogos) == 1:
    print(f"O ano com o maior número de novos jogos é: {ano_mais_jogos[0]}")
else:
    print("Houve um empate entre os seguintes anos:")
    print(ano_mais_jogos)


print('PERGUNTA 3: Qual é a quantidade de jogos por gênero?')
for genre, count in contagem_genero.items():
    print(f"{genre}: {count} jogo(s)")
```
